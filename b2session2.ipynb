```python
import pandas as pd
```

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percentage": [92, 67, 81, 45, 74, 88,],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}
df = pd.DataFrame(data)
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["total_score"] = df["assignment_score"] + df["quiz_score"]
df
```

```python
df["eligible"] = (
    (df["attendance_percentage"] >= 75) &
    (df["total_score"] >= 70) &
    (df["lab_completed"] == True)
)

df
```

```python
summary = {
    "average_attendance": df["attendance_percentage"].mean(),
    "average_score": df["total_score"].mean(),
    "eligible_count": df["eligible"].sum(),
    "total_students": len(df)
}

summary
```

```python
assert df["attendance_percentage"].between(0, 100).all()
assert df["assignment_score"].between(0, 20).all()
assert df["quiz_score"].between(0, 100).all()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(df["student_id"], df["quiz_score"])

plt.title("Student Quiz Scores")
plt.xlabel("Student ID")
plt.ylabel("Quiz Score")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["student_id"], df["attendance_percentage"])

plt.title("Student Attendance")
plt.xlabel("Student ID")
plt.ylabel("Attendance %")

plt.show()
```

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percentage": [92, 67, 81, 45, 74, 88,],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}

df = pd.DataFrame(data)

df
df.shape
df.head()
df.info()

```

```python
# Visualization
import matplotlib.pyplot as plt
plt.bar(df["student_id"], df["quiz_score"])
plt.show()
```

```python
# Create box plot for quiz scores
plt.figure(figsize=(6,5))
plt.boxplot(df["quiz_score"])
plt.title("Quiz Score Box Plot")
plt.ylabel("Quiz Score")

plt.show()
```

```python
# Create pie chart for lab completion status

lab_counts = df["lab_completed"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    lab_counts,
    labels=["Completed", "Not Completed"],
    autopct="%1.1f%%"
)

plt.title("Lab Completion Status")

plt.show()
```

```python
# Create histogram for quiz score distribution

plt.figure(figsize=(8,5))
plt.hist(df["quiz_score"], bins=5)

plt.title("Quiz Score Distribution")
plt.xlabel("Quiz Score")
plt.ylabel("Frequency")

plt.show()
```

```python
# Create scatter plot for attendance and quiz score
plt.figure(figsize=(8,5))

plt.scatter(df["attendance_percentage"], df["quiz_score"])

plt.title("Attendance vs Quiz Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Quiz Score")

plt.show()
```
